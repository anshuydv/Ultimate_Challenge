#### 12000 rows × 11 columns

What makes them an "adopted user"?

Now that it has been determined which users are "adopted users", need to figure out what factors into multiple log-ins into devices.

Some features that will be considered for this model will be:

```python

# 2. drop irrelevant columns, leaving only "object_id", "opted_in_to_mailing_list", "enabled_for_marketing_drip" and 'adopted_user'

users.drop(['creation_time', 'name', 'email', 'creation_source', 'last_session_creation_time', 'org_id', 'invited_by_user_id'], axis=1, inplace=True)
users
```

```python

# 3. set up trainx and trainy

train = users[:9600] #Training Set
train
```

```python
test = users[9600:] #Test Set (based off of Test set being 20% of whole dataset)
test
```

```python
#Setting up Training dataset

trainx = train.drop('adopted_user', axis = 1)

trainy = train['adopted_user']
```

```python

#Setting up Testing Dataset

testx = test.drop('adopted_user', axis = 1)

testy = test['adopted_user']
```

```python

# 4. Feed these into .fit to build models

##.fit to build Logistic Regression (L.R.) Model

from sklearn.linear_model import LogisticRegression

lm = LogisticRegression(class_weight = 'balanced').fit(trainx, trainy)
lm
```

```python
# 5. .predict to predict on training set and test set

#Predicting on Training dataset

lm_train_predictions =  lm.predict(trainx)
lm_train_predictions
```

```python
#LM Training dataset Performance Metrics

from sklearn.metrics import classification_report

report = classification_report(trainy, lm_train_predictions)
print(report)
```

```python
#Predicting on Test dataset

lm_test_predictions = lm.predict(testx)
lm_test_predictions
```

```python

#Logisitc Model (LM) Test dataset Performance Metrics

from sklearn.metrics import classification_report

report = classification_report(testy, lm_test_predictions)
print(report)
```

```python
#.fit to build Decision Tree (DT) model

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

param_grid = [{'max_depth': [1, 2, 3, 4]}]

tree_class = DecisionTreeClassifier()

grid_search = GridSearchCV(tree_class, param_grid, cv = 3, scoring = 'neg_mean_squared_error')

grid_search.fit(trainx, trainy)

#Getting best estimator directly. Inputting results in refit Decision Tree model.
grid_search.best_estimator_
```

```python
#Training DT Model 

#Now with optimized parameters, can .fit model with tuned settings:

tree_class = DecisionTreeClassifier(class_weight = 'balanced', max_depth = 1)
tree_class.fit(trainx,trainy)
```

```python
#Predicting on Training dataset

tree_class_train_predictions = tree_class.predict(trainx)
tree_class_train_predictions
```

```python
#DT Training dataset Performance Metrics

from sklearn.metrics import classification_report

report = classification_report(trainy, tree_class_train_predictions)
print(report)
```

```python
#Predicting on Test dataset

tree_class_test_predictions = tree_class.predict(testx)
tree_class_test_predictions
```

```python
#DT Test dataset Performance Metrics

from sklearn.metrics import classification_report

report = classification_report(testy, tree_class_test_predictions)
print(report)
```

### Conclusions from Summary Tables:

Precision identifies the frequency with which a model was correct when predicting the positive class Recall answers the following question: Out of all the possible positive labels, how many did the model correctly identify? F1-score is the weighted average of the precision and recall, where an F1 score reaches its best value at 1 and worst score at 0.

For Classification model metrics/assessment, values closest to 1 were most desirable signaling correct positive predictions. Model that had mostly 1's was Decision Tree model.

Therefore, to re-assess this data, I would choose the Decision Tree model due to it having a high success rate in predicting which users would be classified as 'adopted user'.
